```python
import matplotlib.pyplot as plt
import torch
```

```python
Energy= torch.load(r"/home/at-lab/ownCloud/bulk_good (1)/Madhav/Energy_Consumed_Per_Episode.pt")
Reward= torch.load(r"/home/at-lab/ownCloud/bulk_good (1)/Madhav/Reward_Per_Episode_1.pt")
Terminal_Reached= torch.load(r"/home/at-lab/ownCloud/bulk_good (1)/Madhav/Terminal_reached.pt")
```

```python
plt.figure(figsize=(20,10))
plt.plot(Energy)
```

```python
plt.figure(figsize=(20,10))

N = 100
cumsum, moving_aves = [0], []

for i, x in enumerate(Energy, 1):
    cumsum.append(cumsum[i-1] + x)
    if i>=N:
        moving_ave = (cumsum[i] - cumsum[i-N])/N
        #can do stuff with moving_ave here
        moving_aves.append(moving_ave)
        
plt.plot(moving_aves)
```
